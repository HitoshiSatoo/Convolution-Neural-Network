## 3. Preparação e exploração dos dados

```python
import os

def walk_through_dir(dir_path):
  """Walks, through_dir_path retorna os conteudos"""
  for dirpath, dirnames, filenames in os.walk(dir_path):
    print(f"Tem {len(dirnames)} diretorios e {len(filenames)} imagens em '{dirpath}")
```
